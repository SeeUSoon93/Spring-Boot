# 1.1 서버와 클라이언트  
인터넷 서비스는 보통 서버와 클라이언트가 관계를 맺는다.


## 1.1.1 클라이언트란?
클라이언트(client)는 서버로 요청하는 프로그램을 모두 일컬어 말함.  
웹 브라우저(클라이언트)에 주소를 입력한 뒤 엔터를 누르는 것이 '서버에 요청'하는 것. 서버는 그 주소에 맞는 화면으로 응답.  

## 1.1.2 서버란?
서버(server)는 클라이언트의 요청을 받아 처리 하는 주체.  
흔히 웹 브라우저에 주소를 입력하는 건 '새로운 화면을 그리기 위한 데이터를 달라'는 데이터 요청에 해당

# 1.2 데이터베이스
데이터베이스(DataBase)는 여러 사람이 한 군데에 모아놓고 여러 사람이 사용할 목적으로 관리하는 데이터 저장소.  
  
클라이언트에서 SQL(DB를 조작하기 위한 언어)로 DB관리 시스템(Oracle, MySQL 등)에 데이터를 요청하면 DB에서 데이터를 꺼내 응답  
  
## 1.2.1 RDB란?
DB는 RDB,NoSQL,NewSQL 등 여러 종류가 있다. 그 중 RDB(Relational DataBase-관계형 데이터베이스)를 가장 많이 사용.  
- 관계형 데이터베이스는 데이터를 행과 열로 이루어진 테이블로 관리. 기본키(primary key)를 사용해 각 행을 식별.
- 각 테이블 간 관계를 지을 수 있음  
  
## 1.2.2 SQL이란?
SQL은 Structured Query Language의 약자. 쿼리, 데이터 검색을 하는 언어. SQL에는 ANSI 표준 SQL이 있고, 각 RDB별 방언이 존재.  
  
## 1.2.3 NoSQL이란?
NoSQL은 SQL을 안 쓴다는 의미이지만, 최근에는 Not Only SQL의 의미로 많이 사용. RDB는 데이터 저장, 질의, 수정, 삭제가 용이하지만 성능을 올리기가 쉽지 않음. DB의 성능을 높이려면 머신의 성능을 좋게 하는 스케일 업 또는 여러 대로 분리하는 스케일 아웃이 필요. 스케일 업은 장비를 업그레이드, 스케일 아웃은 DB를 분산하고 이때 트랜잭션을 사용하면 성능이 떨어짐.  
이러한 문제를 해결하기 위해 NoSQL이 등장. 데이터 모델링을 어떻게 하느냐에 따라 다이나모디비, 카우치베이스, 몽고디비와 같은 DB들이 존재

# 1.3 아이피와 포트
아이피는 인터넷에서 컴퓨터 또는 기기들이 서로를 식별하고 통신하기 위한 주소. 아이피를 알면 서버를 찾을 수 있음. 그러나 서버를 이용하기 위해서는 포트도 알아야함.  
아이피는 서버를 찾기 위한 번호, 포트는 서버에서 운용되고 있는 서비스를 구분하기 위한 번호

# 1.4 라이브러리와 프레임워크

## 1.4.1 라이브러리란?
라이브러리(library)는 애플리케이션 개발에 필요한 기능인 클래스, 함수 등을 모아놓은 코드의 모음.  
프로그래밍에서 라이브러리는 도서관보다는 책들의 모음이라고 이해하는 것이 좋다. 개발자가 소프트웨어를 만들 떄 필요에 따라 원하는 기능을 구현하기 위해 코드의 모음을 가져다 쓸 수 있는 일종의 도구.  
개발자가 직접 복잡한 코드를 작성하지 않아도 되므로 더 빠르게 개발가능
  
## 1.4.2 프레임워크란?
프레임워크는 소프트웨어 개발을 수월하게 하기 위한 소프트웨어 개발 환경. 일(work)하기 위한 틀(frame)을 제공.  
프레임워크는 정해진 틀에서 개발해야 한다는 단점이 있지만 개발 효율이 높다는 장점이 있다.  
  
프레임워크는 애플리케이션을 개발할 때 전체적인 구조를 잡기 위해 사용, 라이브러리는 개발을 하는 과정에서 필요한 기능을 구현하기 위해 사용

# 1.5 백엔드 개발자의 업무
주된 업무는 서버 측 애플리케이션을 개발하는 일


1. 과제할당  
사업부, 마케팅, 개발팀 등에서 개발 과제를 할당. 그 후 개발자가 그 과제를 분석
  
2. 과제 분석  
어떤 소셜 서비스에서 '친구 초대 이벤트'라는 과제가 주어졌다면 '친구'는 무엇인지, '초대'는 무엇을 의미하는지, '이벤트'는 어떤 조건이 필요하고 초대를 한 친구와 초대된 친구에게 보상을 어떻게 줄지, 이벤트를 위해 어떤 데이터들을 저장하는지를 결정해야 함  
또한 제공할 성능과 사용할 기술을 검토하고, 문제가 있다면 해결 방법을 면밀하게 분석하고 정리.
  
3. 개발  
개발 과정에서는 기능을 개발하고, 버그를 수정.  
개발한 코드는 깃허브 등의 프로그램을 사용해 동료의 리뷰를 받고 리뷰에 대한 내용을 반영  
  
4. 테스트  
코드가 제대로 동작하는지 테스트. 로직상 문제는 없는지, 성능 한계는 어느 정도인지, 코드가 읽기 편한지, 클래스/메서드/함수를 다른 사람이 간편하게 사용할 수 있는지 등 검토하면서 테스트.  
개발용 코드보다 테스트를 먼저 작성하는 테스트 주도 개발 기법을 사용하기도 함.  
  
5. QA  
애플리케이션의 품질을 높이기 위한 작업
  
6. 배포와 유지보수  
QA가 완료되면 서버에 배포.  
애플리케이션을 쉽고 안전하게 배포하고, 문제 발생시 쉽게 이전 버전으로 돌리는 작업을 할 수 있게 준비.  
배포 방법
    - 롤링 배포 : 특정 개수의 서버들을 돌아가며 배포
    - 블루 그린 배포 : 똑같은 서버 인스턴스를 띄우고 로드밸런서 설정을 바꾸는 배포
    - 카나리 배포 : 전체 서버의 특정 비율만큼 배포해보고 문제없는 경우 점점 배포를 늘려가는 배포  
   
빌드부터 배포까지의 과정을 자동화하는 것을 CI(continuous integration)/CD(continuous delivery & continuious deployment)라고 함. CI는 지속적인 통합으로 새로운 코드 변경 사항이 정기적으로 빌드 및 테스트되어 저장소에 통합되는 것을 의미. CD는 지속적인 배포.  
CI는 저장소에 코드가 반영되는 것, CD는 실제 프로덕션 환경까지 변경된 코드가 반영되는 것을 의미
