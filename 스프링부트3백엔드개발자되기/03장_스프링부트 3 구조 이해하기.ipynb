# 3.1 스프링 부트3 구조 살펴보기

## 3.1.1 계층

1. 프레젠테이션 계층(Controller) - HTTP 요청을 받고 비즈니스 계층으로 요청을 전송하는 역할.
2. 비즈니스 계층(Service) - 모든 비즈니스 로직(서비스를 만들기 위한 로직)을 처리.
3. 퍼시스턴스 계층(Repository) - 데이터베이스 관련 로직을 처리. 데이터베이스에 접근하는 DAO(DB계층과 상호작용하기 위한 객체)를 사용할 수 있음.

## 3.1.2 디렉터리
1. main - 실제 코드를 작성하는 공간. 프로젝트 실행에 필요한 소스코드, 리소스 파일이 이 폴더 안에 있다.
2. test - 프로젝트의 소스 코드를 테스트할 목적의 코드나 리소스 파일
3. build.gradle - 빌드를 설정하는 파일. 의존성이나 플러그인 설정 등과 같이 빌드에 필요한 설정을 할 때 사용
4. setting.gradle - 빌드할 프로젝트의 정보를 설정하는 파일

## 3.1.3 main 디렉터리 구성하기
1. templates - HTML고ㅓㅏ 같은 뷰 관련 파일
2. static - JS,CSS, 이미지와 같은 정적 파일을 넣을 디렉터리
3. application.yml - 스프링 부트 서버가 실행되면 자동으로 로딩되는 파일. 스프링 부트 설정을 할 수 있는 파일
